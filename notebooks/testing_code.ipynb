# Read JSON data from directory

```python
json_files = sorted([file for file in os.listdir('../tmp/gh_api_data') if file.endswith('.json')])
```

```python
# Initialize an empty dictionary to hold the combined data
combined_data = {}

# Loop over the list of files
for index, js in enumerate(json_files):
    # Construct the full file path
    file_path = os.path.join('../tmp/gh_api_data', js)
    # Read the file
    with open(file_path) as json_file:
        data = json.load(json_file)
    # Add the data to the combined_data dictionary
    combined_data.update(data)
```

```python
data = []

for repo_key, repo_info in combined_data.items():
    # Skip this iteration if repo_info is None or empty
    if not repo_info:
        print(f"Warning: No data for {repo_key}, skipping.")
        continue
    
    # Extract the 'name', 'forkCount', and 'repositoryTopics'
    print(repo_key)
    name = repo_info.get('name', 'Missing')  # Provide a default value in case 'name' is missing
    fork_count = repo_info.get('forkCount', 'Missing')  # Provide a default value in case 'forkCount' is missing
    
    # Safely extract 'repositoryTopics' if they exist
    repo_topics = repo_info.get('repositoryTopics', {}).get('edges', 'Missing')
    topics = [topic['node']['url'].split('/')[-1] for topic in repo_topics if 'node' in topic and 'url' in topic['node']]
    
    # Append a tuple with the extracted information to the data list
    data.append((name, fork_count, topics))
```

```python
df = pd.DataFrame(data, columns=['name', 'forkCount', 'topics'])
```

```python
df['topics'] = df['topics'].apply(lambda x: x if x else "Missing")
```

```python
sub_df = df.query("topics != 'Missing'")
```

```python
import pandas as pd
from itertools import combinations
from collections import defaultdict
# Assuming 'df' is your DataFrame and 'topics' is the column with lists of topics

# Create a default dictionary to count pairs
pair_counts = defaultdict(int)

# Iterate over each list of topics and count pairs
for topics_list in sub_df['topics']:
    # Get all combinations of topics in pairs from the list
    for topic_pair in combinations(topics_list, 2):
        # Sort the pair to ensure consistent ordering (topic1, topic2) == (topic2, topic1)
        sorted_pair = tuple(sorted(topic_pair))
        pair_counts[sorted_pair] += 1

# Now convert the pair counts into a DataFrame
network_df = pd.DataFrame(
    [(pair[0], pair[1], count) for pair, count in pair_counts.items()],
    columns=['topic1', 'topic2', 'edge_count']
)

print(network_df)
```

```python
import dash
import dash_cytoscape as cyto
import dash_html_components as html
import networkx as nx
import community as community_louvain
import matplotlib.pyplot as plt
import random

# 创建网络图
G = nx.karate_club_graph()

# 使用 Louvain 算法检测社区
partition = community_louvain.best_partition(G)

# 为每个节点分配颜色
colors = [random.randint(0, 255) for _ in range(len(set(partition.values())))]
color_map = {community: f'rgb({colors[community]}, {255 - colors[community]}, {colors[community] // 2})' for community in set(partition.values())}

# 转换为 Cytoscape 格式并应用颜色
nodes = [{'data': {'id': str(node), 'label': str(node), 'color': color_map[partition[node]]}} for node in G.nodes]
edges = [{'data': {'source': str(edge[0]), 'target': str(edge[1])}} for edge in G.edges]

# 创建 Dash 应用
app = dash.Dash(__name__)
app.layout = html.Div([
    cyto.Cytoscape(
        id='network-graph',
        elements=nodes + edges,
        style={'width': '100%', 'height': '600px'},
        layout={'name': 'cose'},
        stylesheet=[
            {
                'selector': 'node',
                'style': {
                    'background-color': 'data(color)',
                    'label': 'data(label)'
                }
            }
        ]
    )
])

if __name__ == '__main__':
    app.run_server(debug=True)
```
